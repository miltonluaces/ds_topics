# Tensorflow Keras Functional architectures

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import tensorflow.keras.models as tkm
import tensorflow.keras.layers as tkl
import tensorflow.keras.utils as tku
```

```python
X_train = pd.DataFrame(np.zeros(100))
```

### Concatenate

```python
inp = tkl.Input(shape=X_train.shape[1:], name='inp')
hid1 = tkl.Dense(units=30, activation="relu", name='hid1')(inp)
hid2 = tkl.Dense(units=30, activation="relu", name='hid2')(hid1)
conc = tkl.concatenate([inp, hid2], name='conc')
out = tkl.Dense(units=1, name='out')(conc)
model = tkm.Model(inputs=[inp], outputs=[out])

tku.plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

### Multi-input

```python
inpA = tkl.Input(shape=[5], name='inpWide')
inpB = tkl.Input(shape=[6], name='inpDeep')
hid1 = tkl.Dense(30, activation='relu', name='hid1')(inpB)
hid2 = tkl.Dense(30, activation='relu')(hid1)
conc = tkl.concatenate([inpA, hid2])
out  = tkl.Dense(1, name="output")(conc)
model = tkm.Model(inputs=[inpA, inpB], outputs=[out])

tku.plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

### Multi-output

```python
inpA = tkl.Input(shape=[5], name='inpWide')
inpB = tkl.Input(shape=[6], name='inpDeep')
hid1 = tkl.Dense(30, activation='relu')(inpB)
hid2 = tkl.Dense(30, activation='relu')(hid1)
conc = tkl.concatenate([inpA, hid2])
out1 = tkl.Dense(1, name='out1')(conc)
out2 = tkl.Dense(1, name='out2')(hid2)
model = tkm.Model(inputs=[inpA, inpB], outputs=[out1, out2])

tku.plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```
