# VGG 16

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import os 
from PIL import Image 
import seaborn as sns
import tensorflow.keras.models as tkm
import tensorflow.keras.layers as tkl
import tensorflow.keras.utils as tku
import tensorflow.keras.preprocessing as tkp
import tensorflow.keras.applications as tka
```

### Build model from scratch

```python
input = tkl.Input((224,224,1)) 

conv1  = tkl.Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu")(input)
conv2  = tkl.Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu")(conv1)
pool1  = tkl.MaxPooling2D((2, 2))(conv2)

conv3  = tkl.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu")(pool1)
conv4  = tkl.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu")(conv3)
pool2  = tkl.MaxPooling2D((2, 2))(conv4)

conv5  = tkl.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(pool2)
conv6  = tkl.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(conv5)
conv7  = tkl.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(conv6)
pool3  = tkl.MaxPooling2D((2, 2))(conv7)

conv8  = tkl.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(pool3)
conv9  = tkl.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv8)
conv10 = tkl.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv9)
pool4  = tkl.MaxPooling2D((2, 2))(conv10)

conv11 = tkl.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(pool4)
conv12 = tkl.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv11)
conv13 = tkl.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv12)
pool5  = tkl.MaxPooling2D((2, 2))(conv13)

flat   = tkl.Flatten()(pool5)
dense1 = tkl.Dense(4096, activation="relu")(flat)
dense2 = tkl.Dense(4096, activation="relu")(dense1)
output = tkl.Dense(1000, activation="softmax")(dense2)

model  = tkm.Model(inputs=input, outputs=output)
model.summary()
tku.plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```
