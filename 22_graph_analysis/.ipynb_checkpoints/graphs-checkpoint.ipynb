# Graphs

## Imports

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from warnings import simplefilter
simplefilter(action='ignore')
```

## Load data

```python
data = pd.read_csv('D:/data/csv/data_graph.csv')
data.head()
```

## Create Graph

```python
df = nx.from_pandas_edgelist(data, source='Origin', target='Dest', edge_attr=True)
print(df.nodes())
print(df.edges())
```

## Show Graph

```python
plt.figure(figsize=(12,8))
nx.draw_networkx(df, with_labels=True)
```

## Calculations on graphs

### Shortest path

```python
path = nx.dijkstra_path(df, source='AMA', target='PBI', weight='Distance')
path
```

```python
path = nx.dijkstra_path(df, source='AMA', target='PBI', weight='AirTime')
path
```
