# Ensambles with Mlxtend

```python
from utilities.std_imports import *
import matplotlib.gridspec as gridspec
import itertools
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from mlxtend.data import iris_data

from mlxtend.classifier import EnsembleVoteClassifier
from mlxtend.plotting import plot_decision_regions
```

### Define three classifiers and build the ensemble

```python
clf1 = LogisticRegression(random_state=0)
clf2 = RandomForestClassifier(random_state=0)
clf3 = SVC(random_state=0, probability=True)

eclf = EnsembleVoteClassifier(clfs=[clf1, clf2, clf3], weights=[2, 1, 1], voting='soft')
```

### Testing

```python
X, y = iris_data()
X = X[:,[0, 2]]

gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(10, 8))

models = [clf1, clf2, clf3, eclf]
mod_names = ['Logistic Regression', 'Random Forest', 'SVM', 'Ensemble']
mod_pos = itertools.product([0, 1], repeat=2)

for m, n, p in zip(models, mod_names, mod_pos):
    m.fit(X, y)
    ax = plt.subplot(gs[p[0], p[1]])
    fig = plot_decision_regions(X=X, y=y, clf=m, legend=2)
    plt.title(n)

plt.show()
```
