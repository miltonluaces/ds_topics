```python
import random as rd
import itertools as it
import numpy as np
import pandas as pd
```

```python
v1 = ['a', 'b', 'c', 'd']
v2 = ['A', 'B', 'C']
v3 = [1, 2, 3, 4, 5]
```

```python
pr = list(it.product(v1, v2, v3))
df = pd.DataFrame(pr)
df.columns = ['v1', 'v2', 'v3']
df
```

```python
ages = [16, 18, 33, 38, 53, 44, 61, 55, 25,  19, 15, 40, 52, 33]
df = pd.DataFrame(ages)
df.columns = ['age']
df
```

```python
bins = [15, 30, 40, 50, 60, np.inf]
names = ['15-30', '30-40', '40-50', '50-60', '60+']
df['age_bin'] = pd.cut(df['age'], bins, labels=names)
df
```

```python
bins = [15, 30, 40, 50, 60, np.inf]
df['age_bin'] = pd.cut(df['age'], bins)
df
```
