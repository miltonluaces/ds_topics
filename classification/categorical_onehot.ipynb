# Classification with many categorical variables - One hot encoding

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import tensorflow.keras.models as tkm
import tensorflow.keras.layers as tkl
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

### Load data

```python
ds = pd.read_csv('D:/data/csv/mushrooms.csv')
print(ds.shape)
ds.head()
```

### Analysis

```python
ds.info()
```

```python
levels = pd.DataFrame([ds[c].unique() for c in ds.columns]).transpose().fillna('')
levels.columns = ds.columns
levels
```

```python
y = ds['type']
X = ds.drop(['type'], axis=1)
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### One hot coding

```python
ds_oh = pd.get_dummies(ds)
print(ds_oh.shape)
ds_oh.head()
```

```python
y = ds_oh[['type_e', 'type_p']]
X = ds_oh.drop(['type_e', 'type_p'], axis=1)
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Naive Gaussian

```python
nb = GaussianNB()
nb.fit(X_train, y_train['type_e'])
accurs = cross_val_score(estimator=nb, X=X_train, y=y_train['type_e'], cv=5)  
print(accurs); np.mean(accurs)
```

### SkLearn Random Forest

```python
hypers = {'bootstrap': True,
          'min_samples_leaf': 3,
          'n_estimators': 50, 
          'min_samples_split': 10,
          'max_features': 'sqrt',
          'max_depth': 6,
          'max_leaf_nodes': None}

model_class = RandomForestClassifier(**hypers)
model_class.fit(X_train, y_train)
```

```python
pred_test = model_class.predict(X_test)
```

```python
pred_testDf = pd.DataFrame(pred_test)
pred_testDf.index = y_testDf.index
y_testDf = pd.DataFrame(y_test)
y_testDf.shape, pred_testDf.shape 
```

```python
df_test = pd.concat([y_testDf, pred_testDf], axis=1)
df_test.columns =['real_e', 'real_p', 'pred_e', 'pred_p']
df_test['diff'] = (df_test['pred_e'] < df_test['pred_p']).astype(int)
df_test['hit'] = (df_test['real_p'] == df_test['diff']).astype(int)
print('Accuracy : ', round(df_test['hit'].sum()/df_test.shape[0] * 100,2), '%')
df_test.head()
```

### SkLearn MLP

### Neural networks Tensorflow

```python
model = tkm.Sequential()
model.add(tkl.Dense(8, input_dim=117, activation='relu'))
model.add(tkl.Dense(8, activation='relu'))
model.add(tkl.Dense(2, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=10)
_, accuracy = model.evaluate(X, y, verbose=0)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```
