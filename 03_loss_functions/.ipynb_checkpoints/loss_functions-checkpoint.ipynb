# Loss functions

## Imports and init

```python
import numpy as np
import matplotlib.pyplot as plt

x=range(-100, 100)
y=x; yh = np.zeros(200); 
n=10; d=10
```

## L1 Absolute

$ \large L(y, \hat{y})  = | y - \hat{y} | $

```python
L = abs(y- yh)
plt.plot(L);
```

## L2 Squared

Euclidean, also called squared loss function. Applicable for Regression
Computes the square of the difference between expected and actual values.

$ \large L(y, \hat{y})  = {( y - \hat{y} )}^{2} $

```python
L  = (y- yh)**2
plt.plot(L);
```

## MSE Mean Squared Error

$ \large L(y, \hat{y}) = \frac{1}{n}{( y - \hat{y} )}^{2} $

```python
L = 1/n * (y- yh)**2
plt.plot(L);
```

## Pseudo-Huber

$ \large L_{\delta} (y, \hat{y}) = \begin{cases} 
          \frac{1}{2} {( y - \hat{y} )}^{2} & | y - \hat{y} | \le \delta \\
          \delta | y - \hat{y} | - \frac{1}{2} \delta^2 & otherwise 
       \end{cases} $

```python
L = lambda y, yh : np.select([abs(y-yh)< d, abs(y-yh)>= d], [1/2 * (y-yh)**2 ,  d * abs(y-yh) - 1/2 * d**2 ])
plt.plot(L(y, yh));
```
