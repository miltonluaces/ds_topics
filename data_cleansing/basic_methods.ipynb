### Columns with nulls

Check proportion of nulls in columns

```python
df.isnull().values.sum()

nulls = [int(df.iloc[:,i].isnull().sum() / df.shape[0] *100) for i in range(df.shape[1])]
nulls = pd.DataFrame({'Col': df.columns, 'Nulls': nulls})
```

Drop columns with nulls proportion above a threshold

```python
thres = 90
null_cols = []
for i in range(nulls.shape[0]):
    if nulls.iloc[i,1] >= thres:
        null_cols.append(nulls.iloc[i,0])
        
print(df.shape[1])
df = df.drop(null_cols, axis=1)
print(df.shape[1])
```

### Show results for categorical response

```python
y_test = pd.DataFrame({'a': [1, 0, 0, 0, 0, 0], 'b': [0, 1, 0, 1, 0, 0], 'c': [0, 0, 0, 0, 1, 0], 'd': [0, 0, 1, 0, 0, 0], 'e': [0, 0, 0, 0, 0, 1], })
pred_test = pd.DataFrame({'a': [0, 0, 0, 0, 1, 0], 'b': [0, 1, 0, 1, 0, 0], 'c': [0, 0, 0, 0, 0, 0], 'd': [0, 0, 1, 0, 0, 0], 'e': [0, 0, 0, 0, 0, 1], })
display(y_test)
display(pred_test)
```

```python
def show_results(y_test, pred_test):
    pred_test.index = y_test.index
    res = pd.concat([y_test.add_prefix('d_'), pred_test.add_prefix('p_')], axis=1)
    res['label'] = res[res.columns[:5]].idxmax(axis=1).str[2:]
    res['pred'] = res[res.columns[5:10]].idxmax(axis=1).str[2:]
    res['error'] = (res['label'] != res['pred']).astype(int)
    print('Error : ', round(res['error'].sum() / res.shape[0] * 100, 2), '%')
    return res
```

```python
res = show_results(y_test, pred_test)
res
```

### Replace for categorical variables (for binning)

```python
df = pd.DataFrame({'A': ['a', 'b', 'b', 'c', 'a'], 'B': [1,2,3,4,5]})
df
```

```python
rep = {'a': 'x', 'b': 'y'}
df['A'].replace(rep, inplace=True)
df
```
